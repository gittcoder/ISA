```python
import pandas as pd
import numpy as np
```

```python
# LEVEL 1
df= pd.read_csv('pandas_isatsk2/weatherAUS.csv')
```

```python
df
```

```python
meanMaxTemp=df['MaxTemp'].mean()
```

```python
meanMinTemp=df['MinTemp'].mean()
```

```python
meanMaxTemp=round(meanMaxTemp,1)
```

```python
meanMinTemp=round(meanMinTemp,1)
```

```python
filt=(df['Location']=='Cobar')
```

```python
AvgMaxTemp_Cobar=df.loc[filt]['MaxTemp'].mean()
```

```python
AvgMaxTemp_Cobar=round(AvgMaxTemp_Cobar,1)
import math
```

```python
def count_nan(col):
    count=0
    for i in df[col]:
        if(math.isnan(i)==True):
            count+=1
    return count
```

```python
count_nan_Sunshine=count_nan('Sunshine')
```

```python
count_nan_Evaporation=count_nan('Evaporation')
```

```python
print("Level 1 Result :\nAverage Maximum Temperature = {0}\
\nAverage Minimum Temperature = {1}\nAverage Maximum Temperature in \
the Location Cobar = {2}\nCount of null value in the column Sunshine = {3}\
\nCount of null value in the column Evaporation = {4}"\
     .format(meanMaxTemp,meanMinTemp,AvgMaxTemp_Cobar,count_nan_Sunshine,count_nan_Evaporation))
```
